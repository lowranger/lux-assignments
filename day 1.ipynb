```python
import pandas as pd

df= pd.read_csv(r"simplelinearregression.csv")
df.head(20)
```

```python
#import libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
```

```python
x= df[['Age']]
y= df['Premium']
```

```python
lin =   LinearRegression()
lin.fit(x,y)
```

```python
print("Intercept:", lin.intercept_)
print("Slope:", lin.coef_[0])
```

```python
df['Predicted'] = lin.predict(x)

plt.scatter(df['Age'], df['Premium'], label='Actual', color='blue')
plt.plot(df['Age'], df['Predicted'], label='Regressiopn Line', color='red')
plt.title('Simple Linear Regression')
plt.xlabel('Age')
plt.ylabel('Premium')
plt.legend()
plt.grid()
plt.show()
```
